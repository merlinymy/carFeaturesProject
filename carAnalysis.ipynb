```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dataPath = "cleanedCarFeatures.csv"
```

```python
df = pd.read_csv(dataPath, low_memory = False)
```

```python
#EDA
df.shape
```

```python
plt.figure(figsize=(16,25))
#plt.xticks(rotation=90)
sns.set_theme(style="darkgrid")

ax = sns.countplot(y="Fuel System", data=df, order = df['Fuel System'].value_counts().index, orient="v")
ax.bar_label(ax.containers[0])
plt.savefig('count.png', dpi=300, bbox_inches='tight')
```

```python
plt.figure(figsize=(25,16))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);
# save heatmap as .png file
# dpi - sets the resolution of the saved image in dots/inches
# bbox_inches - when set to 'tight' - does not allow the labels to be cropped
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
```

```python
#subset with only EFI and GDI
subsetEFIandGDI = df[df['Fuel System'].str.contains("Electronic Fuel Injection|Gasoline Direct Injection", na = False)]
```

```python
subsetEFIandGDI.info()
```

```python
plt.figure(figsize=(15,10))
#plt.xticks(rotation=35)
ax = sns.barplot(x="Fuel System", y="Fuel Economy Est-Combined (MPG)", data=subsetEFIandGDI)
plt.savefig('Fuel Economy Est-Combined (MPG).png', dpi=300, bbox_inches='tight')
```

```python
plt.figure(figsize=(15,10))
#plt.xticks(rotation=35)
ax = sns.barplot(x="Fuel System", y="EPA Fuel Economy Est - City (MPG)", data=subsetEFIandGDI)
plt.savefig('EPA Fuel Economy Est - City (MPG).png', dpi=300, bbox_inches='tight')
```

```python
plt.figure(figsize=(15,10))
#plt.xticks(rotation=35)
ax = sns.barplot(x="Fuel System", y="EPA Fuel Economy Est - Hwy (MPG)", data=subsetEFIandGDI)
plt.savefig('EPA Fuel Economy Est - Hwy (MPG)', dpi=300, bbox_inches='tight')
```

```python
plt.figure(figsize=(15,10))
ax = sns.lineplot(x="Year", y="Fuel Economy Est-Combined (MPG)",
             hue="Fuel System",
             data=subsetEFIandGDI)
ax.set_title('Combined Environment')
plt.savefig('Combined Environment timeseries', dpi=300, bbox_inches='tight')
```

```python
plt.figure(figsize=(15,10))
ax = sns.lineplot(x="Year", y="EPA Fuel Economy Est - City (MPG)",
             hue="Fuel System",
             data=subsetEFIandGDI)
ax.set_title('City Environment')
plt.savefig('city Environment timeseries', dpi=300, bbox_inches='tight')
```

```python
plt.figure(figsize=(15,10))
ax = sns.lineplot(x="Year", y="EPA Fuel Economy Est - Hwy (MPG)",
             hue="Fuel System",
             data=subsetEFIandGDI)
ax.set_title('Highway Environment')
plt.savefig('hw Environment timeseries', dpi=300, bbox_inches='tight')
```
